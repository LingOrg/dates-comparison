```python
import numpy as np
import pandas as pd
```

```python
csv_results1 = pd.read_csv('Comparison_Data_1.csv')

csv_results4 = pd.read_csv('Comparison_Data_4.csv')

csv_results5 = pd.read_csv('Comparison_Data_5.csv')
```

```python
csv_results1_null = csv_results1[csv_results1['Rezultati'] == 0]
csv_results4_null = csv_results4[csv_results4['Rezultati'] == 0]
csv_results5_null = csv_results5[csv_results5['Rezultati'] == 0]

print("csv_results1 shape: ", csv_results1_null.shape)
print("csv_results1 accuracy: ", csv_results1.shape[0] - csv_results1_null.shape[0], " / ", csv_results1.shape[0], " = ", (csv_results1.shape[0] - csv_results1_null.shape[0]) / csv_results1.shape[0] * 100, "%" )

print("csv_results4 shape: ", csv_results4_null.shape)
print("csv_results4 accuracy: ", csv_results4.shape[0] - csv_results4_null.shape[0], " / ", csv_results4.shape[0], " = ", (csv_results4.shape[0] - csv_results4_null.shape[0]) / csv_results4.shape[0] * 100, "%" )

print("csv_results5 shape: ", csv_results5_null.shape)
print("csv_results5 accuracy: ", csv_results5.shape[0] - csv_results5_null.shape[0], " / ", csv_results5.shape[0], " = ", (csv_results5.shape[0] - csv_results5_null.shape[0]) / csv_results5.shape[0] * 100, "%" )

combined_null_results = pd.concat([csv_results1_null, csv_results4_null, csv_results5_null])
combined_null_results.to_csv('Combined_Null_Results.csv', index=False)

print("combined_null_results shape: ", combined_null_results.shape)
combined_null_results.head(5)
```
