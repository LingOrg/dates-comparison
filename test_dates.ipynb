```python
import numpy as np
import pandas as pd
```

```python

regulations_df = pd.read_csv('Regulations.csv')

print(regulations_df.shape)

regulations_df.head(2)
```

```python
regulation_versions_df = pd.read_csv('RegulationVersions.csv')
print(regulation_versions_df.shape)

regulation_versions_df.head(2)
```

```python
old_df = pd.read_csv('4_vazeci_OneDate.csv')
print(old_df.shape)

old_df.head(2)
```

```python
corelation_df = pd.read_csv('RegulationRegistarMap.csv')
print(corelation_df.shape)

corelation_df.head(2)
```

#### RegulationVersions by Brojac from 4_vazeci_OneDate.csv


Take Ids that are in old_df and regulation_df

```python
old_df = old_df.rename(columns={'Brojac': 'RegistarId'})
new_df = pd.merge(old_df, corelation_df, on='RegistarId', how='inner')

registarIds_df = new_df[['RegistarId']]
regulationIds_df = new_df[['RegulationId']]
print(registarIds_df.shape)
print(regulationIds_df.shape)
```

Update old_df (remove ids that are not in regulation_df)

```python
old_df = old_df[old_df['RegistarId'].isin(registarIds_df['RegistarId'])]
print(old_df.shape)
old_df.head(2)
```

Take regulations with id from new_df

```python
regulations_df = regulations_df[regulations_df['Id'].isin(regulationIds_df['RegulationId'])]
print(regulations_df.shape)
regulations_df.head(2)
```

#### Find regulationVersions from regulations

```python
regulation_versions_df_temp = regulation_versions_df[regulation_versions_df['RegulationId'].isin(regulationIds_df['RegulationId'])]
regulation_versions_df_temp = regulation_versions_df_temp[regulation_versions_df_temp['VersionNumber'] == 1]
regulation_versions_df_temp['EnterIntoForceDate'] = pd.to_datetime(regulation_versions_df_temp['EnterIntoForceDate'], errors='coerce')

regulation_versions_df_temp['EnterIntoForceDate'] = regulation_versions_df_temp['EnterIntoForceDate'].dt.strftime('%d.%m.%Y').replace('NaT', 'Null')
print(regulation_versions_df_temp.shape)
regulation_versions_df_temp.head(10)
```

# Create dataset for comparison

```python
comparison_df = regulation_versions_df_temp
comparison_df = comparison_df.rename(columns={'Id': 'RegulationId',
                             'EnterIntoForceDate': 'Matko_StupanjeNaSnagu',
                             'ApplicationDate': 'Matko_DatumPrimjene',
                             'EndDate': 'Matko_PrestanakVazenja'})

old_df_temp = old_df[['RegistarId', 'Vazenje', 'Naslov', 'DatumStupanjaNaSnagu', 'DatumPrimjene', 'DatumPrestankaVazenja']]
old_df_temp = old_df_temp.rename(columns={'DatumStupanjaNaSnagu': 'Dario_StupanjeNaSnagu',
                           'DatumPrimjene': 'Dario_DatumPrimjene',
                           'DatumPrestankaVazenja': 'Dario_PrestanakVazenja'})

comparison_df = pd.concat([comparison_df.reset_index(drop=True), old_df_temp.reset_index(drop=True)], axis=1)
#order_of_columns = ['RegulationId', 'RegistarId', 'Naslov', 'Vazenje', 'Matko_StupanjeNaSnagu', 'Dario_StupanjeNaSnagu', 'Matko_DatumPrimjene', 'Dario_DatumPrimjene', 'Matko_PrestanakVazenja', 'Dario_PrestanakVazenja']
order_of_columns = ['RegulationId', 'RegistarId', 'Naslov', 'Vazenje', 'Matko_StupanjeNaSnagu', 'Dario_StupanjeNaSnagu', 'Matko_DatumPrimjene', 'Dario_DatumPrimjene']
comparison_df = comparison_df[order_of_columns]
comparison_df = comparison_df.iloc[:, 1:]

print(comparison_df.shape)
comparison_df.head(2)
```

```python
# Dividing into 5 parts
df_split = np.array_split(comparison_df, 5)
df1 = df_split[0]
df2 = df_split[1]
df3 = df_split[2]
df4 = df_split[3]
df5 = df_split[4]

print(comparison_df.shape)
print(df1.shape)
print(df2.shape)
print(df3.shape)
print(df4.shape)
print(df5.shape)

comparison_df.to_csv('Results/Comparison_Data.csv', index=False)
df1.to_csv('Results/Comparison_Data_1.csv', index=False)
df2.to_csv('Results/Comparison_Data_2.csv', index=False)
df3.to_csv('Results/Comparison_Data_3.csv', index=False)
df4.to_csv('Results/Comparison_Data_4.csv', index=False)
df5.to_csv('Results/Comparison_Data_5.csv', index=False)
```
